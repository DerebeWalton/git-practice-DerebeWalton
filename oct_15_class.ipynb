```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://statistics.byu.edu/directory"
```

```python
r = requests.get(url)
print(r.status_code)
type(r.text)
```

```python
soup = BeautifulSoup(r.text)
type(soup)
```

```python
container = soup.find('main', {'class': 'Page-main'})
items = soup.find_all('div', {'class': 'ListImageOnTopCircular-items-item'})
```

```python
names = []
titles = []
links = []

for item in items: 
    h3 = item.find('h3')
    name = h3.text.strip()
    link = h3.find('a')['href']
    title = item.find('div', {'class': 'PromoImageOnTopCircular-description promo-description'}).text.strip()
    
    names.append(name)
    titles.append(title)
    links.append(link)

df = pd.DataFrame({'name': names, 'title': titles, 'link': links})
df
```
