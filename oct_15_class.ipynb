```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://statistics.byu.edu/directory"
```

```python
r = requests.get(url)
print(r.status_code)
type(r.text)
```

```python
soup = BeautifulSoup(r.text)
type(soup)
```

```python
container = soup.find('main', {'class': 'Page-main'})
items = soup.find_all('div', {'class': 'ListImageOnTopCircular-items-item'})
```

```python
names = []
titles = []
links = []

for item in items: 
    h3 = item.find('h3')
    name = h3.text.strip()
    link = h3.find('a')['href']
    title = item.find('div', {'class': 'PromoImageOnTopCircular-description promo-description'}).text.strip()
    
    names.append(name)
    titles.append(title)
    links.append(link)

df = pd.DataFrame({'name': names, 'title': titles, 'link': links})
df
```

```python
import time
```

```python
websites = []
offices = []
phones = []
emails = []
for link in df['link']:
    time.sleep(1)
    r3 = requests.get(link)
    bs = BeautifulSoup(r3.text)
    
    try:
        website = bs.find('div', {'class': 'EmployeePage-websiteLink'}).find('a')['href']
    except:
        website = None
    websites.append(website)
    try:
        email = bs.find('div', {'class': 'EmployeePage-email description-text'}).find('a').text
    except:
        email = None
    emails.append(email)
    try:
        office = bs.find('div', {'class': 'EmployeePage-address description-text'}).find('div', {'class': 'RichTextModule-items RichTextBody'}).text
    except:
        office = None
    offices.append(office)
    try:
        phone = bs.find('div', {'class': 'EmployeePage-phoneNumber description-text'}).find('a').text
    except:
        phone = None
    phones.append(phone)
```

```python
df = pd.DataFrame({'name': names, 'title': titles,
                   'website': websites, 'email': emails,
                   'office': offices, 'phone': phones})
df
```
